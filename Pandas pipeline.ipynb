### Add col [house_size] based on the number of rooms of column [Avg. Area Number of Rooms]. 
if room <= 4: small
    <4 và <=6 : medium
    >6: big
1. viết 1 loạt điều kiện lọc cho col
2. lồng các điều kiện vào 1 hàm gọi là hàm filter_size
3. apply hàm này cho cột cần lọc.
4. Tạo 1 cột mới = apply (hàm). cột được apply

```python
### Tạo hàm filter_size:
def filter_size (room):
    if room <= 4:
        return "small"
    elif 4 < room <=6:
        return "medium"
    else:
        return "big"
### Apply hàm này cho cột df['Avg. Area Number of Rooms'].apply(filter_size)
### Tạo cột mới df['house_size']
df['house_size'] = df['Avg. Area Number of Rooms'].apply(filter_size)
```

### Function pipe to combine 02 functions for Dataframe

```python
### create fake dataframe
df_pipe = pd.DataFrame()
### create df columns
df_pipe['name']=['Nam','Hoang','Hue','Lan']
df_pipe['age'] = [25, 30, 21, 27]
df_pipe['employed'] = ['Yes','Yes','No','Yes']
```

```python
### create a function to calculate the average age group by employed
def average_age_groupby_employed(df,col):
    return round(df.groupby(col).mean(),1)
### create a function to capital all col
def upper_char (df,col):
    return df.col.upper()
```

```python
### using pipeline to apply to functions:
average_age_groupby_employed(df_pipe,col="employed").pipe(df_pipe.columns="age".upper())
```
